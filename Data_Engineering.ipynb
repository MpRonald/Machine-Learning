# Engenharia de Dados com Python

### Na primeira etapa realizamos a extração dos dados no site:

https://www.b3.com.br/pt_br/market-data-e-indices/servicos-de-dados/market-data/historico/mercado-a-vista/series-historicas/

```python
# imports
import pandas as pd
```

```python
# field's position
cols_pecs = [(2,10),(10,12),(12,24),(27,39),(56,69),(69,82),(82,95),(108,121),
             (152,170), (170,188)]
# field's names
names = ['data_pregao', 'cod_bdi', 'sigla_acao', 'nome_acao', 'preco_abertura',
         'preco_maximo', 'preco_minimo', 'preco_fechamento', 'qtd_negocios',
         'volume_negococios']

# creatring data frame
data = pd.read_fwf('/content/drive/MyDrive/Datasets/COTAHIST_A2022.TXT',
                   colspecs=cols_pecs, names=names, skiprow=1)
data.head()
```

```python
data_std_lot = data[data['cod_bdi'] == 2.0]
data_std_lot
```

```python
data_std_lot.drop(['cod_bdi'], axis=1, inplace=True)
data_std_lot
```

```python
data_std_lot.info()
```

```python
# change to datetime
data_std_lot['data_pregao'] = pd.to_datetime(data_std_lot['data_pregao'], format='%Y%m%d')
data_std_lot.info()
```

```python
data_std_lot.iloc[1:5,3:7] = (data_std_lot.iloc[1:5,3:7] / 100).astype(float)
```
