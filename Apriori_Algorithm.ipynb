```python
# imports
import pandas as pd
from apyori import apriori
```

```python
market_1 = pd.read_csv('https://raw.githubusercontent.com/MpRonald/datasets/main/mercado.csv', header=None)
market_1
```

```python
transaction = []
for i in range(len(market_1)):
  transaction.append([str(market_1.values[i, j]) for j in range(market_1.shape[1])])
```

```python
transaction
```

```python
type(transaction)
```

```python
rules = apriori(transaction, min_support=0.3, min_confidence=0.8, min_lift=2)
results = list(rules)
```

```python
results
```

```python
len(results)
```

```python
results[2][0]
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for result in results:
  #print(resultado)
  s = result[1]
  #print(s)
  result_rules = result[2]
  #print(result_rule)
  for result_rule in result_rules:
    #print(result_rule)
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    #print(a, ' - ', b, ' - ', c, ' - ', l)
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
A
```

```python
B
```

```python
confianca
```

```python
lift
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```

```python
rules_df.sort_values(by = 'lift', ascending = False)
```

```python
trans = []
for i in range(market_2.shape[0]):
  trans.append([str(market_2.values[i, j]) for j in range(market_2.shape[1])])
```

```python
trans
```

```python
rules_2 = apriori(trans, min_support = 0.003, min_confidence = 0.5, min_lift = 3)
results_2 = list(rules_2)
len(results_2)
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for result in results_2:
  #print(resultado)
  s = result[1]
  #print(s)
  result_rules = result[2]
  #print(result_rule)
  for result_rule in result_rules:
    #print(result_rule)
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    #print(a, ' - ', b, ' - ', c, ' - ', l)
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```

```python
rules_df.sort_values(by = 'lift', ascending = False)
```
