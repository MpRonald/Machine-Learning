```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/MpRonald/datasets/main/stress.csv')
data.head()
```

```python
data.columns = [['snoring_rate', 'respiration_rate', 'body_temperature', 'limb_movement',
                'blood_oxygen', 'eye_movement', 'sleeping_hours', 'heart_rate',
                'stress_level']]
```

```python
X = data.drop(['stress_level'], axis=1).values
X
```

```python
y = data['stress_level'].values
y[1:10]
```

```python
scaler = StandardScaler().fit(X)
X = scaler.transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

# SVM

```python
# creating the model
svm = SVC(kernel='linear', random_state=1, C = 1.0)\
.fit(X_train, y_train)
```

```python
# predict
predict = svm.predict(X_test)
```

```python
accuracy_score(y_test, predict)
```

```python
# using 'poly' kernel
svm_credit = SVC(kernel='poly', random_state=1, C = 1.0)\
.fit(X_train, y_train)
predict = svm_credit.predict(X_test)
accuracy_score(y_test, predict)
```

```python
# function to run Grid Search
def run_grid_search(estimators, param, X, y):
  grid = GridSearchCV(estimator=estimators, param_grid=param).fit(X, y)
  print(f"Best Parameter: {grid.best_params_}")
  print(f"Best Index: {grid.best_index_}")
  best_score = round(grid.best_score_,4)
  print(f"Best Score: {best_score}")
```

```python
parameters_S = {
    'tol': [0.1, 0.01, 0.001, 0.0001, 0.00001],
    'C': [1.0, 1.5, 2.0, 2.5, 3.0],
    'kernel':['rbf', 'linear', 'poly', 'sigmoid']}
```

```python
result_svm = run_grid_search(SVC(), parameters_S, X, y)
```

```python
print("Best Parameter: {'C': 1.0, 'kernel': 'rbf', 'tol': 0.1")
```

# Random Forest

```python
# parameters to classifier
parameters_R = {
      'criterion':['gini', 'entropy'],
      'n_estimators':[10, 50, 100, 200],
      'min_samples_split':[2,5,10],
      'min_samples_leaf':[2,6,10]}
```

```python
result_RF = GridSearchCV(RandomForestClassifier(), param_grid = parameters_R).fit(X, y)
best_RF = result_RF.best_score_
best_RF
```
