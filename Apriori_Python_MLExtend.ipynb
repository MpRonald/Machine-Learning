```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('https://github.com/MpRonald/datasets/raw/main/groceries_groceries.csv', header=None)
data = data.iloc[1:,1:]
data
```

```python
dataset = data.replace(np.nan, 0)
dataset
```

## transformando itens em listas e removendo os zeros

```python
def zero_remove(l):
    return list(filter(lambda x: x != 0, l))
```

```python
list_transactions = []

for idx, row in dataset.iterrows():
    transactions = row.values.tolist()
    transactions = zero_remove(transactions)
    list_transactions.append(transactions)
```

```python
list_transactions[0]
```

```python
te = TransactionEncoder()
te_ary = te.fit(list_transactions).transform(list_transactions)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
dataset.head()
```

```python
freq_itemset = apriori(dataset, min_support=.015, use_colnames=True)\
    .sort_values(by='support', ascending=False)
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.45)
rules
```

```python
rules.sort_values(by=['lift'], ascending=False)
```
