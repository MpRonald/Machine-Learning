```
# a instrução % matplotlib embutida foi usada para gerar 
# gráficos dentro do Jupyter Notebook

import matplotlib.pyplot as plt
%matplotlib inline
```

```
import numpy as np
import math
```

```
# utilizando um grafico de linhas
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals]
plt.plot(x_vals, y_vals)
plt.show()
```

No script 1, geramos 20 números equidistantes usando a função np.linspace (). Os números são armazenados na variável x_vals. Em seguida, iteramos por meio de cada valor na lista x_vals e obtemos a raiz quadrada de cada valor. A lista resultante é armazenada na variável y_vals. Para plotar um gráfico de linha por meio do módulo pyplot, você só precisa chamar o método plot() do módulo pyplot e, em seguida, passar os valores para os eixos x e y. É importante mencionar que plt é um apelido para pyplot no script 1 e você pode nomeá-lo como quiser.

```
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals]
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot (x_vals, y_vals)
plt.show()
```

Você pode melhorar a estética e a legibilidade de seus gráficos adicionando títulos, rótulos e legendas a eles. Vamos primeiro ver como adicionar títulos e rótulos a um gráfico.

Para adicionar rótulos em eixos x e y, você precisa passar os valores de cadeia, respectivamente, aos xlabel e ylabel métodos da pyplot módulo.

```
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals]
plt.xlabel ( 'Valores X' )
plt.ylabel ( 'Valores Y' )
plt.title ( 'Square Roots' )
plt.plot (x_vals, y_vals)
plt.show()
```

Além de alterar os títulos e rótulos, você também pode especificar a cor do gráfico de linha. Para fazer isso, basta passar a notação abreviada do nome da cor para a função plot () , por exemplo, “r” para vermelho, “b” para azul e assim por diante.

```
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals]
plt.xlabel ( 'Valores X' )
plt.ylabel ( 'Valores Y' )
plt.title ( 'Square Roots' )
plt.plot (x_vals, y_vals, 'r')
plt.show()
```

Para adicionar uma legenda, você precisa fazer duas alterações. Primeiro, você tem que passar um valor de string para o atributo label da função plot () . Em seguida, você deve passar o valor para o atributo loc do método de legenda do módulo pyplot . No atributo loc , você deve passar a localização de sua legenda. O script a seguir plota uma legenda no canto superior central do gráfico.

```
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals]
plt.xlabel ( 'Valores X' )
plt.ylabel ( 'Valores Y' )
plt.title ( 'Square Roots' )
plt.plot (x_vals, y_vals, 'r', label='Raiz Quadrada')
plt.legend(loc='upper center')
plt.show()
```

Você também pode plotar vários gráficos de linha dentro de um gráfico. Tudo o que você precisa fazer é chamar o método plot () duas vezes com valores diferentes para os eixos x e y.

```
x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt (i) for i in x_vals ** 2]
y2_vals = [math.sqrt (i) for i in x_vals ** 3]
plt.xlabel ( 'Valores X' )
plt.ylabel ( 'Valores Y' )
plt.title ( 'Square Roots' )
plt.plot (x_vals, y_vals, 'r' , label = 'Raiz quadrada' )
plt.plot (x_vals, y2_vals, 'b' , label = 'Cube' )
plt.legend (loc = 'upper center' )
plt.show()
```

# Gráficos de dispersão
Um gráfico de dispersão é usado essencialmente para representar graficamente a relação entre duas colunas numéricas na forma de pontos dispersos. Normalmente, um gráfico espalhado é usado quando, para cada valor no eixo x, existem vários valores no eixo y. Para traçar um gráfico de dispersão, a função scatter() do módulo pyplot é usada. Você deve passar os valores para o eixo x e y. Além disso, você deve passar uma notação abreviada do valor da cor para o parâmetro c. 

```
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal vs Petal Length")
plt.scatter(data["sepal.length"], data["petal.length"], c="b")
plt.show()
```

# Gráfico de Pizza
Os gráficos de pizza, como o nome sugere, exibem a distribuição percentual dos valores em uma coluna categórica em termos de uma função agregada. Por exemplo, o script a seguir mostra a distribuição percentual de cargos em relação às categorias de cargos, ou seja, TI, Marketing, Ciência de Dados e Finanças. Para plotar um gráfico de pizza, o método pie () do módulo pyplot é usado. O primeiro parâmetro é a lista dos valores numéricos que você deseja converter e exibir em porcentagens. Em seguida, você deve passar uma lista de categorias para o parâmetro rótulos. O parâmetro explodir define a magnitude da divisão para cada categoria no gráfico de pizza. O parâmetro autopct define o formato no qual a porcentagem será exibida no gráfico de pizza.

```
labels = 'IT' , 'Marketing' , 'Data Science' , 'Finance'
values = [500, 156, 300, 510]
explode = (0.05, 0.05, 0.05, 0.05)
plt.pie (values, explode = explode, labels = labels,\
         autopct = '% 1.1f %%' , shadow = True)
plt.show()
```

O gráfico 'join' é usado para representar graficamente a distribuição do histograma de duas colunas, uma no eixo x e outra no eixo y. Os seguintes parcelas de script uma trama conjunta para total_bill e ponta colunas do 'tips' dataset.

```
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_data)
```

O gráfico de dispersão pode ser substituído por uma linha de regressão em um gráfico comum. Para fazer isso, você precisa passar reg como o valor para o parâmetro de tipo da função jointplot().

```
sns.jointplot(x = 'size', y = 'total_bill',
              data = tips_data, kind = 'reg' )
```

### Violin Plot

Os gráficos de violino são semelhantes aos gráficos de caixa. No entanto, ao contrário dos gráficos de caixa que representam informações de quartil, os gráficos de violino representam a distribuição geral dos valores nas colunas numéricas. O seguinte roteiro traça duas tramas de violino para os passageiros que viajam sozinhos e para os passageiros que viajam com outro passageiro. A função violinplot() é usada para traçar um enredo de enxame com Seaborn.

```
sns.violinplot(x = 'alone', y = 'age', data = titanic_data)
```

O resultado mostra que, entre os passageiros que viajam sozinhos, são muito poucos os passageiros com menos de 15 anos, como mostra o gráfico do violino laranja à direita. Esse comportamento é compreensível, pois as crianças normalmente estão acompanhadas por alguém. Isso pode ser verificado ainda mais olhando para o gráfico do violino azul à esquerda que corresponde aos passageiros acompanhados por outros passageiros.

# Pandas Bar Plots
Para plotar plotagens de barra com Pandas, você precisa de uma lista de categorias e uma lista de valores. A lista de categorias e a lista de valores devem ter o mesmo comprimento. Vamos traçar um gráfico de barras que mostra a idade média dos passageiros do sexo masculino e feminino.


```
titanic_data.head()
```

```
sex_mean = titanic_data.groupby('sex')['age'].mean ()
```

```
print(sex_mean)
```

```
print(type(sex_mean.tolist()))
```

Em seguida, precisamos criar um novo dataframe Pandas com duas colunas: Gênero e Idade, e então podemos simplesmente usar o método bar() para traçar um gráfico de barra que exibe as idades médias dos passageiros do sexo masculino e feminino no navio do Titanic.

```
df = pd.DataFrame({'Gender':['Female', 'Male'], 'Age':sex_mean.tolist()})
ax = df.plot.bar(x='Gender', y='Age', figsize=(8,6))
```
